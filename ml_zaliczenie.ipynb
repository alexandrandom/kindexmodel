```python
import matplotlib.pyplot as plt
import numpy as np

def draw_line(a,b,x_min=-5, x_max=5):
  x=np.linspace(x_min,x_max, 1000)
  #wzor na funkcję liniową
  y=a * x + b
  plt.plot(x,y, "-r", c=(0.75, 0.1, 0.1, 0.25))
```

```python
lr.coef_, lr.intercept_
data.plot.scatter(x="'KI'", y="'T0'", figsize=(10.0,7.5))
draw_line(lr.coef_, lr.intercept_, x_min=-60, x_max=40)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(data, test_size=0.2)
train_df.shape, test_df.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
#wybieramy zmienną i arkusz odpowiedzi
lr.fit(train_df[["'KI'"]], train_df["'T0'"])

lr.coef_, lr.intercept_
```

Mean absolute error dla zbioru treningowego

```python
from sklearn.metrics import mean_absolute_error
train_pred = lr.predict(train_df[["'KI'"]])
mean_absolute_error(train_df["'T0'"], train_pred)
```

Mean absolute error dla zbioru testowego

```python
#badamy to samo dla danych testowych
test_pred = lr.predict(test_df[["'KI'"]])
mean_absolute_error(test_df["'T0'"], test_pred)
```

MSE - mean squared error dla zbioru treningowego:

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(train_df["'T0'"], train_pred)
```

MSE - mean squared error dla zbioru testowego

```python
mean_squared_error(test_df["'T0'"], test_pred)
```

RMSE - pierwiastek błędu średniokwadratowego dla zbioru treningowego

```python
mean_squared_error(train_df["'T0'"], train_pred, squared=False)
```

RMSE - pierwiastek błędu średniokwadratowego dla zbioru testowego

```python
mean_squared_error(test_df["'T0'"], test_pred, squared=False)
```
